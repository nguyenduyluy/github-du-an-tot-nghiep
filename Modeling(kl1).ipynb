```python
import pandas as pd

ex_filepath ='L:\\New folder\\D·ª∞ √ÅN\\ch√≠nh\\data.xlsx'
df = pd.read_excel(ex_filepath)
print("\nFile: ")
print(df)
```

```python
class DataAnalyzer:
    def __init__(self, dataframe):
        self.df = dataframe

    def count_and_display(self, column_name):
        print(f"C√ÅC LO·∫†I TRONG C·ªòT '{column_name}':\n")
        counts = self.df[column_name].value_counts(dropna=False)
        print(counts)
        print("\n===============================\n")
        return counts

# üëâ S·ª≠ d·ª•ng class:
analyzer = DataAnalyzer(df)

# ƒê·∫øm v√† ph√¢n lo·∫°i c·ªôt 'Employment'
employment_counts = analyzer.count_and_display('Employment')

# ƒê·∫øm v√† ph√¢n lo·∫°i c·ªôt 'RemoteWork'
remotework_counts = analyzer.count_and_display('RemoteWork')
```

```python
# B∆∞·ªõc 1: Nh√≥m d·ªØ li·ªáu theo Employment v√† RemoteWork
grouped = df.groupby(['Employment', 'RemoteWork'])

# B∆∞·ªõc 2: ƒê·∫øm s·ªë l∆∞·ª£ng t·ª´ng t·ªï h·ª£p
employment_remote_count = grouped.size()

# B∆∞·ªõc 3: ƒê∆∞a RemoteWork th√†nh c·ªôt (unstack)
employment_remote_summary = employment_remote_count.unstack(fill_value=0)

# B∆∞·ªõc 4: Hi·ªÉn th·ªã k·∫øt qu·∫£
print("B·∫£ng ph√¢n ph·ªëi gi·ªØa Employment v√† RemoteWork:\n")
print(employment_remote_summary)
```
