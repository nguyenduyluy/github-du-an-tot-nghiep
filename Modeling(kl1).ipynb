```python
import pandas as pd

ex_filepath ='L:\\New folder\\DỰ ÁN\\chính\\data.xlsx'
df = pd.read_excel(ex_filepath)
print("\nFile: ")
print(df)
```

```python
print(df.columns)
```

```python
class DataAnalyzer:
    def __init__(self, dataframe):
        self.df = dataframe

    def count_and_display(self, column_name):
        print(f"CÁC LOẠI TRONG CỘT '{column_name}':\n")
        counts = self.df[column_name].value_counts(dropna=False)
        print(counts)
        print("\n===============================\n")
        return counts

# 👉 Sử dụng class:
analyzer = DataAnalyzer(df)

# Đếm và phân loại cột 'Employment'
employment_counts = analyzer.count_and_display('Employment')

# Đếm và phân loại cột 'RemoteWork'
remotework_counts = analyzer.count_and_display('RemoteWork')
```

```python
# Bước 1: Nhóm dữ liệu theo Employment và RemoteWork
grouped = df.groupby(['Employment', 'RemoteWork'])

# Bước 2: Đếm số lượng từng tổ hợp
employment_remote_count = grouped.size()

# Bước 3: Đưa RemoteWork thành cột (unstack)
employment_remote_summary = employment_remote_count.unstack(fill_value=0)

# Bước 4: Hiển thị kết quả
print("Bảng phân phối giữa Employment và RemoteWork:\n")
print(employment_remote_summary)
```
