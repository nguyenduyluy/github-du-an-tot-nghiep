```python
df['WorkExp'] = df['WorkExp'].astype(float)
df['CompTotal'] = df['CompTotal'].astype(float)
df['JobSat'] = df['JobSat'].astype(float)
df['TimeAnswering'] = df['TimeAnswering'].astype(float)
df['TimeSearching'] = df['TimeSearching'].astype(float)
df['SurveyEase'] = df['SurveyEase'].astype(float)
df['SurveyLength'] = df['SurveyLength'].astype(float)
df['ConvertedCompYearly'] = df['ConvertedCompYearly'].astype(float)
print(df.dtypes)
```

```python
print(df.isnull().sum())
```

```python
#thu nhp trung bnh theo trnh  hc vn
df.groupby("EdLevel")["ConvertedCompYearly"].mean().sort_values(ascending=False).round(2)
```
