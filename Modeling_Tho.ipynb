```python
import pandas as pd
df = pd.read_excel("data.xlsx")
df
```

```python
import pandas as pd

class NhanVienDataModel:
    def __init__(self, data):
        self.data = data

    # 1. Tuổi trung bình theo quốc gia (dù bạn nói nhóm tuổi 25–30, nhưng ta vẫn làm được nếu cột là số)
    def tuoi_theo_quoc_gia_va_hinh_thuc(self):
        return (
            self.data.groupby(['Country', 'Age','RemoteWork'])['ResponseId']
            .count()
            .reset_index()
            .rename(columns={'ResponseId': 'Count'})
            .sort_values(['Count'],ascending =False)
        )

    # 2. Trình độ học vấn phổ biến nhất theo hình thức làm việc
    def trinh_do_hoc_van_pho_bien_theo_hinh_thuc_lam_viec(self):
        return self.data.groupby(['RemoteWork', 'EdLevel'])['ResponseId'] \
                   .count() \
                   .reset_index(name='Count') \
                   .sort_values(['RemoteWork', 'Count'], ascending=[True, False]) \
                   .groupby('RemoteWork') \
                   .head(1) \
                   .reset_index(drop=True)

    # 3. Số năm kinh nghiệm trung bình theo ngành nghề
    def kinh_nghiem_trung_binh_theo_nganh(self):
        df_clean = self.data.copy()
        df_clean['WorkExp'] = pd.to_numeric(df_clean['WorkExp'], errors='coerce')
        return df_clean.groupby('DevType')['WorkExp'].mean().round(1).reset_index().dropna().sort_values(by='WorkExp', ascending=False)
```

```python
model = NhanVienDataModel(df)
print("Tuổi trung bình theo quốc gia và theo hình thức làm việc:")
print(model.tuoi_theo_quoc_gia_va_hinh_thuc().head(10))


print("\nTrình độ học vấn phổ biến nhất theo hình thức làm việc:")
print(model.trinh_do_hoc_van_pho_bien_theo_hinh_thuc_lam_viec())
    
print("\nKinh nghiệm trung bình theo ngành nghề:")
print(model.kinh_nghiem_trung_binh_theo_nganh().head(5))
```
