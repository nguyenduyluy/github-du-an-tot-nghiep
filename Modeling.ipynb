```python
import pandas as pd
df = pd.read_excel("data.xlsx")
df
```

```python
class NhanVienDataModel:
    def __init__(self, data):
        self.data = data

    def tinh_luong_trung_binh_theo_remote(self):
        ket_qua = self.data.groupby('RemoteWork')['CompTotal'].mean().round(2)
        return ket_qua.apply(lambda x: f"${x:,.2f} USD")
    
    def so_luong_theo_hinh_thuc(self):
        return self.data['RemoteWork'].value_counts().reset_index(name='SoLuong')

    def ti_le_remote(self):
        tong = len(self.data)
        phan_bo = self.data['RemoteWork'].value_counts(normalize=True) * 100
        return phan_bo.round(2).astype(str) + '%'

    def phan_loai_hai_long(self):
        conditions = [-1, 4, 7, 10]
        labels = ['Khng hi lng', 'Trung bnh', 'Hi lng']
        self.data['MucHaiLong'] = pd.cut(self.data['JobSat'], bins=conditions, labels=labels)
        return self.data[['JobSat', 'MucHaiLong']].head()

    def luong_trung_binh_quoc_gia_va_hinh_thuc(self):
        # Tnh lng trung bnh
        result = self.data.groupby(['Country', 'RemoteWork'])['CompTotal'].mean().reset_index()
        result = result.round(2)

        # nh dng lng theo USD d c
        result['CompTotal'] = result['CompTotal'].apply(lambda x: f"${x:,.2f} USD")
        return result

```

```python
model = NhanVienDataModel(df)

print("Lng trung bnh theo hnh thc lm vic:")
print(model.tinh_luong_trung_binh_theo_remote())

print("\n Tng quan s lng theo hnh thc lm vic:")
print(model.so_luong_theo_hinh_thuc())

print("\nT l % tng hnh thc lm vic:")
print(model.ti_le_remote())

print("\nPhn loi hi lng:")
print(model.phan_loai_hai_long())

bang_luong = model.luong_trung_binh_quoc_gia_va_hinh_thuc()
print("\nLng trung bnh theo quc gia v hnh thc lm vic:")
print(bang_luong.head(10))
```

```python
import pandas as pd

class HieuSuatLamViecModel:
    def __init__(self, data):
        self.data = data
    def ty_le_su_dung_ai(self):
        df_filtered = self.data[self.data['AISelect'] == 'Yes']
        result = df_filtered['RemoteWork'].value_counts(normalize=True).reset_index(name='TyLeDungAI')
        result['TyLeDungAI'] = (result['TyLeDungAI'] * 100).round(2).astype(str) + '%'  # Chuyn sang % v lm trn

        return result
        
    def kinh_nghiem_trung_binh(self):
        result = self.data.groupby('RemoteWork')['WorkExp'].mean().reset_index(name='KinhNghiemTB')
        result['KinhNghiemTB'] = result['KinhNghiemTB'].round(2)
        return result
    
    def muc_do_tin_tuong_ai(self):
        return self.data.groupby(['RemoteWork', 'AIAcc'])['ResponseId'].count().reset_index(name='SoNguoi')
    
    def ty_le_remote_theo_trinh_do(self):
        df = self.data[['RemoteWork', 'EdLevel']].dropna()
        result = df.value_counts(normalize=True).reset_index(name='TyLe')
        result['TyLe'] = (result['TyLe'] * 100).round(2).astype(str) + '%'
        return result.sort_values(by='TyLe', ascending=False)

    # 2. Loi cng vic ph bin theo tng hnh thc lm vic
    def cong_viec_pho_bien_theo_remote(self):
        df = self.data[['RemoteWork', 'DevType']].dropna()
        result = df.groupby('RemoteWork')['DevType'].agg(lambda x: x.mode().iat[0]).reset_index(name='CongViecPhoBien')
        return result
```

```python
model = HieuSuatLamViecModel(df)

print('\nT l s dng AI:')
print(model.ty_le_su_dung_ai())

print('\nKinh nghim trung bnh theo hnh thc:')
print(model.kinh_nghiem_trung_binh())

print('\nMc  tin tng AI ')
print(model.muc_do_tin_tuong_ai())

print("T l hnh thc lm vic theo trnh  hc vn:")
print(model.ty_le_remote_theo_trinh_do())

print("\nCng vic ph bin nht theo hnh thc lm vic:")
print(model.cong_viec_pho_bien_theo_remote())
```
